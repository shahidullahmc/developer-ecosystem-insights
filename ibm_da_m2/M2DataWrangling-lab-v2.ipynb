### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


````python
# Identify inconsistent or irrelevant entries in the 'Country' column

# View unique country names
print("Unique Country Values:\n")
print(df['Country'].unique())

# Count how many times each country appears
print("\nCountry Value Counts:\n")
print(df['Country'].value_counts())

# Identify obviously inconsistent or irrelevant entries
# (e.g., blank strings, 'Unknown', 'Other', or malformed values)
print("\nPotentially Inconsistent Country Entries:\n")
print(df['Country'][df['Country'].isin(['Unknown', 'Other', '', 'NA'])])
````

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


````python
# Step 3.2: Standardize entries in 'Country' and 'EdLevel'

# Example mappings for Country
country_mapping = {
    'United States of America': 'USA',
    'United States': 'USA',
    'US': 'USA',
    'U.S.A.': 'USA',
    'UK': 'United Kingdom',
    'U.K.': 'United Kingdom',
    'England': 'United Kingdom',
    'Viet Nam': 'Vietnam',
    'Russian Federation': 'Russia',
    'Other': None,
    'Unknown': None,
    'NA': None,
    'N/A': None
}

df['Country'] = df['Country'].replace(country_mapping)

# Example mappings for EdLevel
edlevel_mapping = {
    "Bachelor’s degree": "Bachelors",
    "Bachelor's degree": "Bachelors",
    "Bachelors degree": "Bachelors",
    "Master’s degree": "Masters",
    "Master's degree": "Masters",
    "Some college/university study without earning a degree": "Some College",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "High School",
    "Primary/elementary school": "Primary School",
    "Professional degree": "Professional",
    "Other doctoral degree": "Doctorate",
    "NA": None,
    "N/A": None
}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Verify the cleaned categories
print(df['Country'].value_counts(dropna=False).head(20))
print(df['EdLevel'].value_counts(dropna=False).head(20))
````

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


````python
## Write your code here
# One‑hot encode the Employment column
employment_dummies = pd.get_dummies(df['Employment'], prefix='Employment')

# Add encoded columns to the dataframe
df = pd.concat([df, employment_dummies], axis=1)

# Preview the updated dataframe
df.head()
````

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


````python
## Write your code here
# Identify columns with the highest number of missing values
missing_values = df.isnull().sum().sort_values(ascending=False)

print("Missing Values by Column (Highest to Lowest):\n")
print(missing_values)
````

**Note**
> This step calculates how many missing values each column contains and sorts them from highest to lowest. Columns with the largest number of missing entries are the most important to address first, since they may affect later steps such as encoding, normalization, or modeling. Understanding the missing‑value pattern helps decide whether to drop columns, impute values, or apply more advanced cleaning strategies.

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


````python
# Impute missing values in numerical columns

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Impute using median (recommended for skewed data like compensation)
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Verify that missing values are handled
df[numeric_cols].isnull().sum()
````

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


````python
# Impute missing values in categorical columns using the most frequent value (mode)
# Select only categorical (object or string) columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Impute each categorical column with its mode
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Verify that missing values are handled
df[categorical_cols].isnull().sum()
````

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>

````python
## Write your code here
# Apply Min‑Max Scaling to normalize ConvertedCompYearly

df['ConvertedCompYearly_MinMax'] = (
    (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) /
    (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
)

# Preview the result
df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head()
````

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


````python
import numpy as np

# Log‑transform the ConvertedCompYearly column to reduce skewness
# Add 1 to avoid issues with zero or negative values
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Preview the transformed values
df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head()

# ## --Without Numpy, Using Panda
# # Log‑transform without NumPy
# df['ConvertedCompYearly_Log'] = df['ConvertedCompYearly'].apply(lambda x: pd.np.log1p(x))
````

**Note** 
>The log transform reduces the impact of extremely large salary values by compressing the right tail of the distribution. Using np.log1p() ensures the transformation works safely even when values are zero.

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


````python
# Create ExperienceLevel from YearsCodePro

# Convert YearsCodePro to numeric (handles strings like "Less than 1 year")
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

def experience_bucket(x):
    if pd.isna(x):
        return "Unknown"
    if x < 3:
        return "Beginner"
    if x < 7:
        return "Intermediate"
    return "Expert"

df['ExperienceLevel'] = df['YearsCodePro'].apply(experience_bucket)

df[['YearsCodePro', 'ExperienceLevel']].head()
````
