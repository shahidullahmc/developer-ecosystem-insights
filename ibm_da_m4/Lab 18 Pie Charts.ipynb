### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
# Step 1: Select the relevant column
db = df['DatabaseWantToWorkWith'].dropna()

# Step 2: Split multi-select entries
db_expanded = db.str.split(';').explode().str.strip()

# Step 3: Count database frequencies
db_counts = db_expanded.value_counts().head(5)

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(
    db_counts,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)

plt.title('Top 5 Databases Respondents Want to Work With')

# Add legend on the right
plt.legend(
    patches,
    db_counts.index,
    title="Databases",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.show()

# # for simple view
# plt.figure(figsize=(8, 8))
# plt.pie(
#     db_counts,
#     labels=db_counts.index,
#     autopct='%1.1f%%',
#     startangle=140,
#     colors=plt.cm.Set3.colors
# )

# plt.title('Top 5 Databases Respondents Want to Work With')
# plt.show()
```

- `Summary of the Result`

The pie chart highlights the top five databases that respondents most want to work with. Each slice represents one database, and the size of the slice reflects how frequently it was selected. PostgreSQL takes the largest portion, showing that it is the most admired and desired database among developers in this survey. SQLite, MySQL, MongoDB, and Redis follow, each holding a meaningful share of interest. The distribution shows a mix of both traditional and modern database technologies, suggesting that developers value reliability, flexibility, and performance across different use cases. Overall, the chart provides a quick snapshot of current preferences in the developer community and reveals which databases are shaping future interest.

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# Step 1: Select the relevant column
dev = df['DevType'].dropna()

# Step 2: Split multi-select entries
dev_expanded = dev.str.split(';').explode().str.strip()

# Step 3: Count developer role frequencies
dev_counts = dev_expanded.value_counts().head(5)

# Step 4: Plot the pie chart with legend on the right
plt.figure(figsize=(8, 8))
patches, texts, autotexts = plt.pie(
    dev_counts,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors
)

plt.title('Top 5 Developer Roles Among Respondents')

# Legend on the right
plt.legend(
    patches,
    dev_counts.index,
    title="Developer Roles",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.show()
```
