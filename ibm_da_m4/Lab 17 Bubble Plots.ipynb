### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Prepare data
df_bubble = df[['Age', 'SOPartFreq', 'JobSat']].copy()

# Mapping
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

participation_map = {
    'Multiple times per day': 6,
    'Daily or almost daily': 5,
    'A few times per week': 4,
    'A few times per month or weekly': 3,
    'Less than once per month or monthly': 2,
    'I have never participated in Q&A on Stack Overflow': 1
}

df_bubble['AgeNumeric'] = df_bubble['Age'].map(age_map)
df_bubble['FreqNumeric'] = df_bubble['SOPartFreq'].map(participation_map)

# Drop missing
df_bubble = df_bubble.dropna(subset=['AgeNumeric', 'FreqNumeric', 'JobSat'])

# Bubble size
df_bubble['BubbleSize'] = df_bubble['JobSat'] * 20

# Plot
plt.figure(figsize=(12, 7))
plt.scatter(
    df_bubble['AgeNumeric'],
    df_bubble['FreqNumeric'],
    s=df_bubble['BubbleSize'],
    alpha=0.4,
    c=df_bubble['JobSat'],
    cmap='viridis'
)

plt.title('Age vs Participation Frequency (Bubble Size = Job Satisfaction)')
plt.xlabel('Age (Numeric Midpoint)')
plt.ylabel('Participation Frequency (1 = Never, 6 = Multiple times/day)')
plt.colorbar(label='Job Satisfaction')
plt.grid(axis='both', linestyle='--', alpha=0.3)

plt.show()
```

#### Interpretation Summary
>The bubble chart gives you a three‑layer view of how people of different ages participate on Stack Overflow and how satisfied they are with their jobs. Age is shown along the x‑axis, and participation frequency rises up the y‑axis, so you can quickly see how often each age group engages. The bubble sizes and colors reflect job satisfaction, which makes it easy to spot where higher or lower satisfaction clusters. Most age groups show a wide spread of participation habits, and satisfaction varies inside each band rather than following a single pattern. The chart highlights that engagement levels don’t rise or fall consistently with age, and job satisfaction appears to be influenced by factors beyond age or participation alone.

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Prepare data
df_bubble2 = df[['ConvertedCompYearly', 'JobSat', 'Age']].copy()

# Convert numeric fields
df_bubble2['ConvertedCompYearly'] = pd.to_numeric(df_bubble2['ConvertedCompYearly'], errors='coerce')
# Optional: remove extreme outliers for readability
df_bubble2['ConvertedCompYearly'] = df_bubble2['ConvertedCompYearly'].where(df_bubble2['ConvertedCompYearly'] < 300000)

df_bubble2['JobSat'] = pd.to_numeric(df_bubble2['JobSat'], errors='coerce')

# Age mapping (same as Task 1)
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_bubble2['AgeNumeric'] = df_bubble2['Age'].map(age_map)

# Drop missing values
df_bubble2 = df_bubble2.dropna(subset=['ConvertedCompYearly', 'JobSat', 'AgeNumeric'])

# Bubble size scaled for visibility
df_bubble2['BubbleSize'] = df_bubble2['AgeNumeric'] * 5

# Plot
plt.figure(figsize=(12, 7))
plt.scatter(
    df_bubble2['ConvertedCompYearly'],
    df_bubble2['JobSat'],
    s=df_bubble2['BubbleSize'],
    alpha=0.4,
    c=df_bubble2['AgeNumeric'],
    cmap='plasma'
)

plt.title('Job Satisfaction vs Compensation (Bubble Size = Age)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (0–10)')
plt.colorbar(label='Age (Numeric Midpoint)')
plt.grid(axis='both', linestyle='--', alpha=0.3)

plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_lang = df[['Age', 'LanguageHaveWorkedWith']].copy()

# Step 2: Drop missing values
df_lang = df_lang.dropna()

# Step 3: Split languages into separate rows
df_lang = df_lang.assign(
    Language=df_lang['LanguageHaveWorkedWith'].str.split(';')
).explode('Language')

# Step 4: Clean whitespace
df_lang['Language'] = df_lang['Language'].str.strip()

# Step 5: Count frequency of each language per age group
lang_counts = df_lang.groupby(['Age', 'Language']).size().reset_index(name='Count')

# Step 6: Map Age to numeric midpoints (same mapping as before)
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

lang_counts['AgeNumeric'] = lang_counts['Age'].map(age_map)
lang_counts = lang_counts.dropna(subset=['AgeNumeric'])

# Step 7: Limit to top 10 most common languages for readability
top_langs = lang_counts.groupby('Language')['Count'].sum().nlargest(10).index
lang_counts = lang_counts[lang_counts['Language'].isin(top_langs)]

# Step 8: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    lang_counts['AgeNumeric'],
    lang_counts['Language'],
    s=lang_counts['Count'] * 1,   # bubble size
    alpha=0.5,
    c=lang_counts['Count'],
    cmap='viridis'
)

plt.title('Technology Preferences by Age (Bubble Size = Frequency)')
plt.xlabel('Age (Numeric Midpoint)')
plt.ylabel('Programming Language')
plt.colorbar(label='Frequency')
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_db = df[['DatabaseWantToWorkWith', 'JobSat']].copy()

# Step 2: Drop missing values
df_db = df_db.dropna(subset=['DatabaseWantToWorkWith', 'JobSat'])

# Step 3: Convert JobSat to numeric
df_db['JobSat'] = pd.to_numeric(df_db['JobSat'], errors='coerce')
df_db = df_db.dropna(subset=['JobSat'])

# Step 4: Split multi-select database column
df_db = df_db.assign(
    Database=df_db['DatabaseWantToWorkWith'].str.split(';')
).explode('Database')

# Step 5: Clean whitespace
df_db['Database'] = df_db['Database'].str.strip()

# Step 6: Count frequency of each database
db_counts = df_db.groupby('Database').size().reset_index(name='Count')

# Step 7: Compute average JobSat per database
db_avg = df_db.groupby('Database')['JobSat'].mean().reset_index()

# Step 8: Merge counts + satisfaction
db_plot = pd.merge(db_counts, db_avg, on='Database')

# Step 9: Limit to top 10 databases for readability
top_dbs = db_plot.nlargest(10, 'Count')
top_dbs['BubbleSize'] = top_dbs['Count'] * 2

# Step 10: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    top_dbs['JobSat'],
    top_dbs['Database'],
    s=top_dbs['BubbleSize'],
    alpha=0.5,
    c=top_dbs['Count'],
    cmap='viridis'
)

plt.title('Preferred Databases vs Job Satisfaction (Bubble Size = Respondent Count)')
plt.xlabel('Average Job Satisfaction (0–10)')
plt.ylabel('Database')
plt.colorbar(label='Respondent Count')
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.show()
```

**Conclusion Summary**

Across all the bubble plots, the patterns make it clear that people’s choices in tools and technologies carry different levels of popularity and satisfaction, but none of these factors operate in isolation. Age, compensation, participation habits, and technology preferences each show their own spread rather than a single predictable trend. The visualizations help highlight where large groups cluster and where smaller but meaningful pockets of respondents stand out. Together, these charts give a more rounded view of how developers work, what they prefer, and how those choices connect to their overall experience in the field.
>This bubble chart shows how preferred databases relate to average job satisfaction among respondents. Each database sits on the y‑axis, while the x‑axis reflects the typical satisfaction level reported by people who want to work with that technology. The bubble sizes make it easy to see which databases attract the largest number of respondents, and the color scale reinforces those differences. The plot highlights both popularity and sentiment at the same time, showing that some widely used databases cluster around similar satisfaction levels while others stand out with higher or lower averages. It’s a quick way to compare how database preferences align with overall job satisfaction across the survey group.

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_roles = df[['DevType', 'ConvertedCompYearly', 'JobSat']].copy()

# Step 2: Clean numeric fields
df_roles['ConvertedCompYearly'] = pd.to_numeric(df_roles['ConvertedCompYearly'], errors='coerce')
# Optional: remove extreme outliers for readability
df_roles['ConvertedCompYearly'] = df_roles['ConvertedCompYearly'].where(df_roles['ConvertedCompYearly'] < 300000)

df_roles['JobSat'] = pd.to_numeric(df_roles['JobSat'], errors='coerce')

# Step 3: Drop missing values
df_roles = df_roles.dropna(subset=['DevType', 'ConvertedCompYearly', 'JobSat'])

# Step 4: Split multi-select DevType column
df_roles = df_roles.assign(
    Role=df_roles['DevType'].str.split(';')
).explode('Role')

# Step 5: Clean whitespace
df_roles['Role'] = df_roles['Role'].str.strip()

# Step 6: Compute average compensation per role
role_comp = df_roles.groupby('Role')['ConvertedCompYearly'].mean().reset_index()

# Step 7: Compute average job satisfaction per role
role_sat = df_roles.groupby('Role')['JobSat'].mean().reset_index()

# Step 8: Count respondents per role
role_count = df_roles.groupby('Role').size().reset_index(name='Count')

# Step 9: Merge all metrics
role_plot = role_comp.merge(role_sat, on='Role').merge(role_count, on='Role')

# Step 10: Limit to top 12 roles for readability
top_roles = role_plot.nlargest(12, 'Count')
top_roles['BubbleSize'] = top_roles['JobSat'] * 40  # bubble size based on satisfaction

# Step 11: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    top_roles['ConvertedCompYearly'],
    top_roles['Role'],
    s=top_roles['BubbleSize'],
    alpha=0.5,
    c=top_roles['JobSat'],
    cmap='viridis'
)

plt.title('Compensation Across Developer Roles (Bubble Size = Job Satisfaction)')
plt.xlabel('Average Yearly Compensation (USD)')
plt.ylabel('Developer Role')
plt.colorbar(label='Average Job Satisfaction')
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.show()
```

`Interpretation Summary`

This bubble chart compares compensation across different developer roles and shows how satisfaction varies within those groups. Each role sits on the y‑axis, while the x‑axis reflects the typical yearly compensation earned by people in that role. The bubbles help bring the picture together: larger and brighter bubbles indicate higher job satisfaction, making it easy to see which roles combine strong pay with a more positive day‑to‑day experience. The chart highlights that compensation alone doesn’t tell the full story—some roles with similar pay levels show different satisfaction patterns, and others stand out with higher averages on both fronts.

>The dataset contains many respondents for each developer role, each with their own compensation and satisfaction values. Using the **average** gives a single, representative number for each role, making the comparison clearer and preventing the plot from being overwhelmed by thousands of individual points. It smooths out noise, reduces the impact of outliers, and allows the bubble chart to focus on overall trends rather than individual variations.

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_tools = df[['NEWCollabToolsHaveWorkedWith', 'Age']].copy()

# Step 2: Drop missing values
df_tools = df_tools.dropna(subset=['NEWCollabToolsHaveWorkedWith', 'Age'])

# Step 3: Split multi‑select collaboration tools
df_tools = df_tools.assign(
    Tool=df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';')
).explode('Tool')

# Step 4: Clean whitespace
df_tools['Tool'] = df_tools['Tool'].str.strip()

# Step 5: Map age groups to numeric midpoints
age_map = {
    'Under 18 years old': 15,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df_tools['AgeNumeric'] = df_tools['Age'].map(age_map)
df_tools = df_tools.dropna(subset=['AgeNumeric'])

# Step 6: Count tool usage per age group
tool_counts = df_tools.groupby(['Tool', 'AgeNumeric']).size().reset_index(name='Count')

# Step 7: Limit to top 10 most used tools
top_tools = tool_counts.groupby('Tool')['Count'].sum().nlargest(10).index
tool_counts = tool_counts[tool_counts['Tool'].isin(top_tools)]

# Step 8: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    tool_counts['AgeNumeric'],
    tool_counts['Tool'],
    s=tool_counts['Count'] * 2,
    alpha=0.5,
    c=tool_counts['Count'],
    cmap='viridis'
)

plt.title('Collaboration Tools by Age (Bubble Size = Usage Frequency)')
plt.xlabel('Age (Numeric Midpoint)')
plt.ylabel('Collaboration Tool')
plt.colorbar(label='Usage Frequency')
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.show()
```

`Explaining Summary`

The bubble chart highlights how collaboration tool usage varies across different age groups. Each tool appears along the y‑axis, while the x‑axis represents the numeric midpoint of each age category. The size and color of the bubbles reflect how frequently each tool is used, making it easy to spot which tools dominate across the developer community.

>A few patterns stand out. Tools like Visual Studio Code and Visual Studio show consistently large bubbles across most age ranges, suggesting broad adoption regardless of age. Others, such as Neovim or Sublime Text, show more concentrated usage within specific age bands, indicating that certain tools appeal more strongly to particular generations. The visualization helps reveal how collaboration habits shift with age and which tools maintain strong, cross‑generational appeal.

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_web = df[['WebframeWantToWorkWith', 'JobSat']].copy()

# Step 2: Drop missing values
df_web = df_web.dropna(subset=['WebframeWantToWorkWith', 'JobSat'])

# Step 3: Convert JobSat to numeric
df_web['JobSat'] = pd.to_numeric(df_web['JobSat'], errors='coerce')
df_web = df_web.dropna(subset=['JobSat'])

# Step 4: Split multi-select frameworks
df_web = df_web.assign(
    Framework=df_web['WebframeWantToWorkWith'].str.split(';')
).explode('Framework')

# Step 5: Clean whitespace
df_web['Framework'] = df_web['Framework'].str.strip()

# Step 6: Count respondents per framework
fw_counts = df_web.groupby('Framework').size().reset_index(name='Count')

# Step 7: Compute average JobSat per framework
fw_avg = df_web.groupby('Framework')['JobSat'].mean().reset_index()

# Step 8: Merge counts + satisfaction
fw_plot = fw_counts.merge(fw_avg, on='Framework')

# Step 9: Limit to top 10 frameworks for readability
top_fw = fw_plot.nlargest(10, 'Count')
top_fw['BubbleSize'] = top_fw['Count'] * 2   # scaled for visibility

# Step 10: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    top_fw['JobSat'],
    top_fw['Framework'],
    s=top_fw['BubbleSize'],
    alpha=0.5,
    c=top_fw['Count'],
    cmap='viridis'
)

plt.title('Preferred Web Frameworks vs Job Satisfaction (Bubble Size = Respondent Count)')
plt.xlabel('Average Job Satisfaction (0–10)')
plt.ylabel('Web Framework')
plt.colorbar(label='Respondent Count')
plt.grid(axis='x', linestyle='--', alpha=0.3)

plt.show()
```

`Explaining Interpretation of the Output`

The bubble chart highlights how different web frameworks relate to job satisfaction among developers. Each framework appears along the y‑axis, while the x‑axis shows the average satisfaction score reported by respondents who prefer working with that framework. The bubble sizes represent how many people selected each framework, and the color scale reinforces those usage levels.

>A few patterns stand out. Popular frameworks like React, Node.js, and Next.js show large bubbles, indicating strong adoption across the community. Frameworks such as Svelte and FastAPI appear with smaller bubbles but slightly higher satisfaction averages, suggesting that while they have smaller user bases, the developers who choose them tend to report a more positive experience. Meanwhile, widely used frameworks like Angular and Express cluster around similar satisfaction levels, showing steady but not exceptional sentiment. Overall, the visualization helps compare both popularity and satisfaction at a glance, revealing how developer preferences align with their day‑to‑day experience using these tools.
>>Job satisfaction is measured on a numeric scale (0–10).
Each framework is selected by many respondents, and each respondent has their own satisfaction score. If you plotted every individual point, the chart would be unreadable.

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Step 1: Keep needed columns
df_admire = df[['LanguageAdmired', 'Country']].copy()

# Step 2: Drop missing values
df_admire = df_admire.dropna(subset=['LanguageAdmired', 'Country'])

# Step 3: Split multi-select admired technologies
df_admire = df_admire.assign(
    Tech=df_admire['LanguageAdmired'].str.split(';')
).explode('Tech')

# Step 4: Clean whitespace
df_admire['Tech'] = df_admire['Tech'].str.strip()

# Step 5: Count admiration frequency per country
tech_counts = df_admire.groupby(['Country', 'Tech']).size().reset_index(name='Count')

# Step 6: Limit to top 10 admired technologies globally
top_techs = tech_counts.groupby('Tech')['Count'].sum().nlargest(10).index
tech_counts = tech_counts[tech_counts['Tech'].isin(top_techs)]

# Optional: Limit to top 20 countries
top_countries = tech_counts.groupby('Country')['Count'].sum().nlargest(20).index
tech_counts = tech_counts[tech_counts['Country'].isin(top_countries)]

# Step 7: Bubble Plot
plt.figure(figsize=(14, 8))
plt.scatter(
    tech_counts['Country'],
    tech_counts['Tech'],
    s=tech_counts['Count'] * 3,
    alpha=0.5,
    c=tech_counts['Count'],
    cmap='viridis'
)

plt.title('Admired Technologies Across Countries (Bubble Size = Admiration Frequency)')
plt.xlabel('Country')
plt.ylabel('Technology')
plt.xticks(rotation=45, ha='right')
plt.colorbar(label='Admiration Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

`InterpretationSummary of the Output`

The bubble chart illustrates how admiration for different programming technologies varies across countries. Each technology appears along the y‑axis, while the x‑axis lists the countries included in the analysis. The size of each bubble represents how frequently respondents from a given country selected that technology as one they admire, and the color scale reinforces those frequency levels.

>A few patterns emerge clearly. Some technologies—such as Python, JavaScript, and SQL—show large bubbles across many countries, indicating broad global admiration and strong international communities. Others, like Rust or TypeScript, appear with more concentrated clusters, suggesting that enthusiasm for these technologies is strong but varies more by region. The visualization helps highlight both global trends and regional differences, showing which technologies have widespread appeal and which ones resonate more strongly within specific countries.
