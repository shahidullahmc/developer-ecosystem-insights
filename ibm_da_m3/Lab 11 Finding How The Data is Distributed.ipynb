### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
# # #Examine the structure of the data
# 1. Display the number of rows and columns (Shape)
print("Dataset Shape:", df.shape)

# 2. Display the column names to see available variables
print("\nColumn Names:")
print(df.columns.tolist())

# 3. Get detailed information about data types and non-null counts
print("\nData Structure and Types:")
df.info()

# 4. Generate summary statistics for numerical columns
print("\nSummary Statistics (Numerical):")
display(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Handle Missing Data

# Count missing values in each column
missing_counts = df.isna().sum().sort_values(ascending=False)
print("Missing values per column:\n")
print(missing_counts)
```

```python
# Handle categorical columns
categorical_cols = df.select_dtypes(include=['object', 'string']).columns
df[categorical_cols] = df[categorical_cols].fillna('Unknown')

# Leave numeric columns as-is (NaN is correct for EDA)

# Verify
print("\nMissing values after cleaning:\n")
print(df.isna().sum().sort_values(ascending=False))
```

```python
print("Remaining missing values in dataset:", df.isnull().sum().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
# Analyze Key Columns

print("Employment distribution:\n")
print(df['Employment'].value_counts(dropna=False))

print("\nJob Satisfaction distribution:\n")
print(df['JobSat'].value_counts(dropna=False))

print("\nProfessional Coding Experience distribution:\n")
print(df['YearsCodePro'].value_counts(dropna=False))
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
plt.figure(figsize=(10, 10))

df['JobSat'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colormap='viridis'
)

plt.title('Job Satisfaction Distribution')
plt.ylabel('')
plt.show()
```

### Interpretation

The interpretation depends on your actual output, but the typical pattern in this dataset is:

- A large portion of respondents fall under **“Unknown”** because many skipped the question.
- Among those who answered, the most common categories tend to be **“Satisfied”** or **“Very satisfied”**.
- Lower satisfaction categories (e.g., “Dissatisfied”) usually appear in smaller proportions.
- The distribution is often right‑skewed toward positive satisfaction.

### Job Satisfaction Interpretation

The distribution is strongly concentrated in the mid‑to‑high satisfaction range. The largest slices are **7.0 (≈22%)** and **8.0 (≈26%)**, showing that most respondents report being generally satisfied with their jobs. Levels **6.0**, **9.0**, and **10.0** also form substantial portions, reinforcing the overall positive trend. Lower satisfaction levels—1.0, 2.0, 3.0, and 4.0—appear only in small percentages, indicating that very few respondents report being unhappy. The pattern suggests a workforce that is broadly content, with dissatisfaction being relatively rare.

1) Many developers who responded tend to report **moderate to high satisfaction**.
2) Missing responses are substantial, so any conclusions should consider that a large group did not report satisfaction.
3) The dataset suggests a generally positive sentiment among those who answered.

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
# Split and explode both columns
df_lang = df.copy()

df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_lang['LanguageWantToWorkWith'] = df_lang['LanguageWantToWorkWith'].str.split(';')

have = df_lang.explode('LanguageHaveWorkedWith')
want = df_lang.explode('LanguageWantToWorkWith')

# Count frequencies
have_counts = have['LanguageHaveWorkedWith'].value_counts()
want_counts = want['LanguageWantToWorkWith'].value_counts()

# Combine into one table
lang_compare = pd.DataFrame({
    'HaveWorkedWith': have_counts,
    'WantToWorkWith': want_counts
}).fillna(0)

# Keep top 15 languages by total interest
lang_compare['Total'] = lang_compare['HaveWorkedWith'] + lang_compare['WantToWorkWith']
lang_compare = lang_compare.sort_values('Total', ascending=False).head(15)
lang_compare = lang_compare.drop(columns='Total')

lang_compare
```

```python
plt.figure(figsize=(14, 7))
lang_compare.plot(kind='bar', figsize=(14, 7))

plt.title('Programming Languages: Worked With vs Want To Work With')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

### Programming Languages Summary

The comparison between **LanguageHaveWorkedWith** and **LanguageWantToWorkWith** shows a clear divide between established, widely used languages and those gaining momentum. JavaScript, Python, SQL, and HTML/CSS dominate current usage and also remain highly desired, reflecting their central role in modern development. TypeScript, Go, and Rust stand out with much higher “want to work with” counts than “have worked with,” signaling strong forward interest and growing adoption potential. In contrast, languages such as PHP, C, and Java show higher usage than interest, suggesting they are important in existing systems but less preferred for future work. Overall, the distribution highlights a shift toward modern, expressive, and performance‑oriented languages while maintaining strong reliance on foundational technologies.

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
# 1. Keep only rows with both country and remote-work data
df_remote = df.dropna(subset=['Country', 'RemoteWork'])

top10 = df_remote['Country'].value_counts().head(10).index
df_remote = df_remote[df_remote['Country'].isin(top10)]

# 2. Group and count combinations
remote_counts = df_remote.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')

# 3. Pivot for grouped bar chart
pivot_remote = remote_counts.pivot(index='Country', columns='RemoteWork', values='Count').fillna(0)

# 4. Plot
plt.figure(figsize=(14, 7))
pivot_remote.plot(kind='bar', figsize=(14, 7), edgecolor='black')

plt.title('Remote Work Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Remote Work Trends — Top 10 Countries Heatmap

# 1. Keep only rows with both country and remote-work data
df_remote = df.dropna(subset=['Country', 'RemoteWork'])

# 2. Count respondents per country
country_counts = df_remote['Country'].value_counts().head(10).index

# 3. Filter to top 10 countries
df_top = df_remote[df_remote['Country'].isin(country_counts)]

# 4. Group and count combinations
remote_counts = (
    df_top
    .groupby(['Country', 'RemoteWork'])
    .size()
    .reset_index(name='Count')
)

# 5. Pivot into matrix form
pivot_remote = (
    remote_counts
    .pivot(index='Country', columns='RemoteWork', values='Count')
    .fillna(0)
)

# 6. Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    pivot_remote,
    cmap='viridis',
    annot=True,
    fmt='.0f',
    linewidths=0.5,
    linecolor='gray'
)

plt.title('Remote Work Distribution (Top 10 Countries)')
plt.xlabel('Remote Work Category')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Conclusion

Remote‑work preferences across the top respondent countries show clear geographic differences. The United States leads in fully remote roles, reflecting strong adoption of flexible work models. Germany and the United Kingdom show substantial hybrid participation, indicating balanced workplace expectations. Countries like India display a more even spread across all categories, suggesting diverse organizational practices. Ukraine shows a strong remote‑first pattern, while several European countries lean toward hybrid arrangements. These contrasts highlight how national work cultures, infrastructure, and external conditions shape remote‑work trends globally.

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
# Correlation between Job Satisfaction and Experience

# 1. Clean YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 50,
    'Unknown': None
})

# 2. Keep only needed columns
df_corr = df[['JobSat', 'YearsCodePro']].dropna()

# 3. Convert both to numeric
df_corr['JobSat'] = pd.to_numeric(df_corr['JobSat'], errors='coerce')
df_corr['YearsCodePro'] = pd.to_numeric(df_corr['YearsCodePro'], errors='coerce')

# 4. Drop any remaining NaN
df_corr = df_corr.dropna()

# 5. Pearson correlation (works without SciPy)
pearson_corr = df_corr['JobSat'].corr(df_corr['YearsCodePro'], method='pearson')

print("Pearson correlation:", pearson_corr)
print("Pearson correlation:", pearson_corr)

if pearson_corr > 0:
    print("Interpretation: Job satisfaction increases slightly as experience increases.")
elif pearson_corr < 0:
    print("Interpretation: Job satisfaction decreases slightly as experience increases.")
else:
    print("Interpretation: No relationship between job satisfaction and experience.")
```

### Conclusion Takeaway
>The correlation of 0.10 shows a very weak positive relationship, meaning job satisfaction tends to rise only slightly as professional coding experience increases.

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Cross-tabulation Analysis (Employment vs. Education Level)

# 1. Create the cross-tab
ct = pd.crosstab(df['Employment'], df['EdLevel'])

print("Cross‑tabulation (Employment vs. Education Level):\n")
display(ct)

# 2. Improved stacked bar plot with manual margin control
fig, ax = plt.subplots(figsize=(20, 10))

ct.plot(
    kind='bar',
    stacked=True,
    ax=ax,
    colormap='tab20',
    edgecolor='none'
)

ax.set_title('Employment Status by Education Level', fontsize=18, pad=20)
ax.set_xlabel('Employment Status', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

ax.set_xticklabels(ax.get_xticklabels(), rotation=25, ha='right', fontsize=12)
ax.tick_params(axis='y', labelsize=12)

ax.legend(
    title='Education Level',
    bbox_to_anchor=(1.15, 1),
    loc='upper left',
    fontsize=10,
    title_fontsize=11
)

ax.grid(axis='y', linestyle='--', alpha=0.35)

# Manual margin adjustment (this removes the warning)
plt.subplots_adjust(bottom=0.25, top=0.90, right=0.80)

plt.show()



# # 1. Cross‑tab
# ct = pd.crosstab(df['Employment'], df['EdLevel'])

# display(ct)

# # 2. Plot with improved readability
# plt.figure(figsize=(14,7))
# ct.plot(kind='bar', stacked=True, figsize=(14,7), colormap='tab20')

# plt.title('Employment Status by Education Level')
# plt.xlabel('Employment Status')
# plt.ylabel('Count')

# # Rotate x‑labels for clarity
# plt.xticks(rotation=45, ha='right')

# # Move legend outside
# plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')

# plt.tight_layout()
# plt.show()
```

### ***Note***
The cross‑tabulation shows a clear pattern: full‑time employment is concentrated among respondents with **Bachelor’s** and **Master’s degrees**, while categories such as **secondary school**, **primary school**, or **non‑degree pathways** appear more frequently in part‑time, mixed, or student employment groups. This distribution suggests that higher formal education is strongly associated with stable, full‑time roles, whereas lower or non‑traditional education levels correspond to more varied or transitional employment statuses.
